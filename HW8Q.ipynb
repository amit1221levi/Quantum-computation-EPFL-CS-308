Import libraries and set up the simulator



```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, Aer, transpile
from qiskit.visualization import plot_histogram
from IPython.display import display
qreg =  QuantumRegister(2)
creg = ClassicalRegister(2)

```

Define the circuits


```python

#Circuit 1
circuit1 = QuantumCircuit(qreg,creg)
circuit1.h(0)
circuit1.h(1)
circuit1.cx(0, 1)
circuit1.h(1)
circuit1.measure(qreg,creg)


#Circuit 2
circuit2 = QuantumCircuit(qreg, creg)
circuit2.h(0)
circuit2.measure(0, 0)
circuit2.x(1).c_if(creg, 1)
circuit2.h(1)
circuit2.measure(1, 1)

#Circuit 3

circuit3 = QuantumCircuit(1, 1)
circuit3.h(0)
circuit3.x(0)
circuit3.h(0)
circuit3.measure(0, 0)


```
